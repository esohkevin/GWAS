```python
import datetime
import pandas as pd
import numpy as np
import csv
import sys
import vcf
import subprocess
import os

for i in range(0,22):
    chr = i+1
    in_file = "chr%s_pos.txt" % chr 
    out_file = "tchr%s-phased_wref.locs" % chr            #f'chr{chr}-phased_wref.locs'

    with open(in_file, "r+") as inFile:
        pos_file = pd.read_csv(inFile, sep='\t', header=None)
        pos_file1 = pos_file[1]
        pos_file1a = pos_file1/1000                            #pos_file1/1000000
        len_pos_file1 = len(pos_file1)
        pos_len = (max(pos_file1)-min(pos_file1))/1000         #(max(pos_file1)-min(pos_file1))/1000000
        loc_head = pd.DataFrame([(len_pos_file1,pos_len,"L")])
        
    with open(out_file, "w+") as outFile:
        loc_head.to_csv(outFile, header=False, index=False, sep='\t')
        outFile.close()
        
    with open(out_file, "a") as outFile:
        pos_file1a.to_csv(outFile, header=False, index=False, sep='\t')
        outFile.close()
        
    print(loc_head, "\n")
        
```
